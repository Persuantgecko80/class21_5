# From Linear Regression to Machine Learning

```python
# import libraries
import numpy as np
import pandas as pd
```

## Import CSV Data from github

```python
url='https://raw.githubusercontent.com/amir188dor/operations24/main/Salary_Data.csv'
df = pd.read_csv(url)
df.head(5)

# Dataset= pd.read_csv('Salary_Data_Grade.csv')
# Dataset.head()
```

```python
# prompt: split df to x that will be yearsexperience and y that is salary

x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
print(x)
print(y)
```

```python
# Some Examples

#[1]- only line one
#[1:20] - 1 to 20 (without 20)
#[:20] - all lines till 20 (without 20)
#[1:] - from line 1 to the end
#[:] - all



```

# Splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=0)


print(y_test)
print(x_test)
```

## Train

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
reg.intercept_
```

```python
reg.coef_
```

```python
# prompt: what is the code to predict the salary for 10 years expirience

y_pred = reg.predict([[10]])
print(y_pred)
```

```python
# Forecast
print(reg.predict([[25]]))
```

```python
reg.predict(x_test)
```

```python
print(y_test)
```

```python
# compare y_test to predictaion of y_test with x_test

from sklearn.metrics import r2_score
r2_score(y_test,reg.predict(x_test))
```
